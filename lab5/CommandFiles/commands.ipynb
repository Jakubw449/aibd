```python
import pandas as pd

df = pd.read_csv('../OriginalData/weather.txt')

# przesunięcie pierwszego rzędu z tytułu kolumny na pierwszy rząd

df = (df.T.reset_index().T.reset_index(drop=True)
            .set_axis([f'Column 1'], axis=1))

# rozdzielenie ciągu cyfr na konkretne wartości

new_df = pd.DataFrame()
new_df['station_id'] = df['Column 1'].map(lambda x: x[:2]) + df['Column 1'].map(lambda x: x[6:11])
new_df['year'] = df['Column 1'].map(lambda x: x[11:15])
new_df['month'] = df['Column 1'].map(lambda x: x[15:17])
new_df['variables'] = df['Column 1'].map(lambda x: x[17:21])

# rozdzielenie temperatur oraz zamiana ich na float oraz ustawnieie przecinka w dobrym miejscu

for i in range(1,32):
    if i == 1:
        new_df[f'Day {i}'] = df['Column 1'].map(lambda x: x[23:27]).apply(lambda x: float(x.split()[0].replace('0D', '0')))/10
    else: 
        new_df[f'Day {i}'] = df['Column 1'].map(lambda x: x[23+(8*(i-2)):27+(8*(i-2))]).apply(lambda x: float(x.split()[0].replace('D', '0').replace('B', '0')))/10

# wybranie wartości z 2010 roku

new_df = new_df.loc[new_df['year'] == '2010']

# zmiana wartosci '-9999' na brak wartosci

new_df = new_df.replace(99.9, '-', regex=True)

new_df.to_csv('../AnalysisData/SortedData.csv', index=False)

new_df
```
