```python
# Jakub Wąsik ćwiczenie 3

import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-1, 1)
x2 = np.linspace(-6, 6)
x3 = np.linspace(0, 5)

y1 = x1**2 + 5
y2 = x2**2 + 5
y3 = x3**2 + 5

def plotting(a, b):
    plt.plot(a, b)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('wykres f(x)')
    plt.legend('x')
    plt.show()

plotting(x1, y1)
plotting(x2, y2)
plotting(x3, y3)
```
